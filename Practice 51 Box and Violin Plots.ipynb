```python
import numpy as np
import pandas as pd
from numpy.random import randn

from scipy import stats
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
data1 = randn(100)
data2 = randn(100)
```

```python
[data1,data2]
```

```python
sns.boxplot([data1, data2])
```

```python
sns.boxplot([data1, data2], whis = np.inf)
```

```python
sns.boxplot([data1,data2], vert = True)
```

```python
#normal distribution
data1 = stats.norm(0, 5).rvs(100)
#Two gamma dist concatenated together
```

```python
data2 = np.concatenate([stats.gamma(5).rvs(50),-1*stats.gamma(5).rvs(50)])
```

```python
#box plot both data1 and data2
sns.boxplot([data1],vert = True)
```

```python
sns.boxplot(data2, vert = True)
```
